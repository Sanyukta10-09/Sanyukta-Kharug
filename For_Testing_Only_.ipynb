```python
import os
import requests
import matplotlib.pyplot as plt
import pyinputplus as pyip
import datetime # Needed for date handling in visualisations
import matplotlib.dates # Needed for date formatting on plot axis

# --- Setup and Configuration ---

# Get API key from environment, or prompt user if not set
api_key = os.environ.get("OPENWEATHER_API_KEY")
if not api_key:
    print("OpenWeatherMap API key not found in environment variables.")
    try:
        api_key = pyip.inputStr("Please enter your OpenWeatherMap API key: ")
        os.environ['OPENWEATHER_API_KEY'] = api_key # Set for this session
    except Exception as e:
        print(f"Failed to get API key input: {e}")
        api_key = None # Ensure api_key is None if input fails


# --- Weather Data Function ---

def get_weather_data(forecast_days=5):
    """
    Retrieves current and forecast weather data for Port Louis, Mauritius using OpenWeatherMap API.

    Parameters:
        forecast_days (int): Number of forecast days to retrieve (maximum 5 for free tier).

    Returns:
        dict: {
            'current': {'summary': str},
            'forecast': [
                {
                    'date': str,
                    'high': float,
                    'low': float,
                    'precipChance': int,
                    'summary': str
                },
                ...
            ]
        }
        or a dict with an 'error' key if something went wrong.
    """
    # Get API key from environment (set previously or via input)
    api_key = os.environ.get("OPENWEATHER_API_KEY")

    if not api_key:
        return {"error": "‚ùå Configuration error: OpenWeatherMap API key is not set."}

    try:
        # üìç Fixed location
        location = "Port Louis, Mauritius"

        # üåê Prepare API request
        # Using the forecast endpoint which provides 5 days of 3-hour data
        url = (
            f"https://api.openweathermap.org/data/2.5/forecast"
            f"?q={location}&appid={api_key}&units=metric"
        )

        response = requests.get(url)
        response.raise_for_status() # Raises an HTTPError for bad responses (4xx or 5xx)
        data = response.json()

        if not data or not data.get("list"):
             return {"error": "‚ùå API returned no data or invalid format."}

        # üìä Build forecast list - extracting one entry per day (the first one)
        forecast_data = []
        seen_dates = set()

        # The forecast data is in 3-hour intervals. We'll take the first entry
        # for each new day encountered up to the desired number of forecast_days.
        for entry in data.get("list", []):
            # Extract date string (YYYY-MM-DD)
            dt_txt = entry.get("dt_txt")
            if not dt_txt:
                continue # Skip entries without a datetime

            date = dt_txt.split(" ")[0]

            # Process this entry if it's a new day and we haven't reached forecast_days
            if date not in seen_dates and len(forecast_data) < forecast_days:
                seen_dates.add(date)

                # Extract required information, providing default values in case keys are missing
                temp_main = entry.get("main", {})
                weather_info = entry.get("weather", [{}])[0] # weather is a list, take the first element

                forecast_entry = {
                    "date": date,
                    # Note: temp_max and temp_min from the forecast endpoint at
                    # the 3-hour interval might not be the true daily max/min.
                    # This extracts the temp_max/min from the first entry of the day.
                    "high": round(temp_main.get("temp_max", 0.0), 1),
                    "low": round(temp_main.get("temp_min", 0.0), 1),
                    # 'pop' is probability of precipitation
                    "precipChance": round(entry.get("pop", 0.0) * 100),
                    "summary": weather_info.get("description", "N/A").capitalize()
                }
                forecast_data.append(forecast_entry)

            # Stop if we've collected enough days
            if len(forecast_data) >= forecast_days:
                break

        # üå•Ô∏è Current condition - Use the weather description from the very first entry
        # as a proxy for the current condition.
        # Use .get() for safety
        current_summary = data.get("list", [{}])[0].get("weather", [{}])[0].get("description", "N/A").capitalize()
        current = {"summary": current_summary}


        return {
            "current": current,
            "forecast": forecast_data
        }

    except requests.exceptions.RequestException as e:
        # Catch specific requests exceptions for network/API call issues
        return {"error": f"‚ùå Network or API request error: {e}"}
    except ValueError as e:
         # Catch value errors, like issues with data processing
         return {"error": f"‚ùå Data processing error: {e}"}
    except KeyError as e:
        # Catch errors if expected keys are missing in the JSON response
        return {"error": f"‚ùå Data format error: Missing key {e}"}
    except Exception as e:
        # Catch any other unexpected errors
        return {"error": f"‚ùå An unexpected error occurred: {e}"}


# --- Visualisation Functions ---

def create_temperature_visualisation(weather_data):
    """
    Creates and displays a temperature forecast visualisation (High and Low).

    Parameters:
        weather_data (dict): The dictionary returned by get_weather_data().
                             Expected format: {'forecast': [{'date': str, 'high': float, 'low': float, ...}, ...]}

    Returns:
        None. Displays a matplotlib plot.
    """
    if "error" in weather_data or not weather_data.get("forecast"):
        print("Cannot create temperature visualisation: No valid forecast data.")
        return

    forecast = weather_data["forecast"]

    # Extract data for plotting, ensuring defaults
    dates = [day.get("date", "N/A") for day in forecast]
    highs = [day.get("high", 0.0) for day in forecast]
    lows = [day.get("low", 0.0) for day in forecast]

    # Convert date strings to datetime objects for better plotting on the x-axis
    plot_dates = []
    plot_highs = []
    plot_lows = []
    for i, d in enumerate(dates):
        if d != 'N/A':
            try:
                plot_dates.append(datetime.datetime.strptime(d, "%Y-%m-%d"))
                plot_highs.append(highs[i])
                plot_lows.append(lows[i])
            except ValueError:
                print(f"Warning: Could not parse date '{d}'. Skipping entry for visualisation.")
                continue # Skip this date if parsing fails


    if not plot_dates:
        print("No valid dates found in forecast data for visualisation.")
        return

    # Create the plot
    plt.figure(figsize=(10, 6)) # Set the figure size

    # Plot high temperatures
    plt.plot(plot_dates, plot_highs, marker='o', linestyle='-', color='red', label='High Temp (¬∞C)')

    # Plot low temperatures
    plt.plot(plot_dates, plot_lows, marker='o', linestyle='-', color='blue', label='Low Temp (¬∞C)')

    # Add titles and labels
    plt.title("Temperature Forecast (Port Louis)")
    plt.xlabel("Date")
    plt.ylabel("Temperature (¬∞C)")
    plt.legend()
    plt.grid(True)

    # Format x-axis dates
    plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(matplotlib.dates.DayLocator())
    plt.xticks(rotation=45, ha='right') # Rotate dates for better readability

    plt.tight_layout() # Adjust layout to prevent labels overlapping

    # Show the plot
    plt.show()


def create_precipitation_visualisation(weather_data):
    """
    Creates and displays a precipitation chance forecast visualisation.

    Parameters:
        weather_data (dict): The dictionary returned by get_weather_data().
                             Expected format: {'forecast': [{'date': str, 'precipChance': int, ...}, ...]}

    Returns:
        None. Displays a matplotlib plot.
    """
    if "error" in weather_data or not weather_data.get("forecast"):
        print("Cannot create precipitation visualisation: No valid forecast data.")
        return

    forecast = weather_data["forecast"]

    # Extract data for plotting, ensuring defaults
    dates = [day.get("date", "N/A") for day in forecast]
    precip_chances = [day.get("precipChance", 0) for day in forecast]

    # Convert date strings to datetime objects for better plotting on the x-axis
    plot_dates = []
    plot_precip_chances = []
    for i, d in enumerate(dates):
         if d != 'N/A':
            try:
                plot_dates.append(datetime.datetime.strptime(d, "%Y-%m-%d"))
                plot_precip_chances.append(precip_chances[i])
            except ValueError:
                print(f"Warning: Could not parse date '{d}'. Skipping entry for visualisation.")
                continue # Skip this date if parsing fails


    if not plot_dates:
        print("No valid dates found in forecast data for visualisation.")
        return

    # Create the plot
    plt.figure(figsize=(10, 6)) # Set the figure size

    # Create a bar chart for precipitation chance
    # We need positions for the bars and labels
    x_positions = range(len(plot_dates))
    # Use plot_dates directly with bar for better date handling on x-axis
    plt.bar(plot_dates, plot_precip_chances, color='skyblue', width=0.6) # Add width for bar spacing

    # Add titles and labels
    plt.title("Precipitation Chance Forecast (Port Louis)")
    plt.xlabel("Date")
    plt.ylabel("Precipitation Chance (%)")
    plt.ylim(0, 100) # Precipitation chance is between 0 and 100
    plt.grid(axis='y', linestyle='--') # Add horizontal grid lines

    # Format x-axis dates
    plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(matplotlib.dates.DayLocator())
    plt.xticks(rotation=45, ha='right') # Rotate dates for better readability

    plt.tight_layout() # Adjust layout

    # Show the plot
    plt.show()


# --- Main Application Logic (Example Usage) ---

# Call the function to get weather data
weather_data = get_weather_data(forecast_days=5) # Specify the number of forecast days

# Display the results
if "error" in weather_data:
    print(weather_data["error"])
else:
    print("üìç Current Weather in Port Louis:", weather_data.get("current", {}).get("summary", "N/A"))
    # Safely get forecast length
    forecast_length = len(weather_data.get('forecast', []))
    print(f"üìÖ {forecast_length}-Day Forecast:")
    if weather_data.get("forecast"):
        for day in weather_data["forecast"]:
            print(f"  {day.get('date', 'N/A')} ‚Äî üå°Ô∏è {day.get('low', 'N/A')}¬∞C to {day.get('high', 'N/A')}¬∞C | ‚òî {day.get('precipChance', 'N/A')}% | {day.get('summary', 'N/A')}")
    else:
        print("  No forecast data available.")

    # Create and display visualisations if data is available
    print("\nCreating visualisations...")
    create_temperature_visualisation(weather_data)
    create_precipitation_visualisation(weather_data)
```

```python
# Install necessary libraries
# This command is specific to Jupyter Notebook/IPython and runs a shell command
try:
    import pyinputplus as pyip
except ImportError:
    print("Installing pyinputplus...")
    !pip install pyinputplus
    import pyinputplus as pyip # Import after installation

# Install dateutil if needed
try:
    from dateutil.parser import parse as parse_date
    from dateutil.relativedelta import relativedelta
except ImportError:
    print("Installing python-dateutil...")
    !pip install python-dateutil
    from dateutil.parser import parse as parse_date
    from dateutil.relativedelta import relativedelta


# Import other necessary libraries
import os
import requests
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates


# --- Setup and Configuration ---

# Get API key from environment, or prompt user if not set
# The user will be prompted to enter their key if it's not already set
api_key = os.environ.get("OPENWEATHER_API_KEY")
if not api_key:
    print("OpenWeatherMap API key not found in environment variables.")
    try:
        # Use pyinputplus to securely get the API key from the user
        # Added a timeout in case input doesn't work in some environments
        api_key = pyip.inputStr("Please enter your OpenWeatherMap API key: ", timeout=60)
        # Set the environment variable for the current session
        os.environ['OPENWEATHER_API_KEY'] = api_key
        print("API key successfully set for this session.")
    except pyip.TimeoutException:
        print("Input timed out. No API key entered.")
        api_key = None
    except Exception as e:
        print(f"Failed to get API key input: {e}. Please ensure you have pyinputplus installed.")
        api_key = None # Ensure api_key is None if input fails

# Check again if api_key was successfully obtained
if not api_key:
    print("‚ùå Error: OpenWeatherMap API key is not available. Cannot proceed with weather queries.")


# --- Weather Data Function ---

def get_weather_data(location, forecast_days=5):
    """
    Retrieves current and forecast weather data for a specified location
    using OpenWeatherMap API.
    ... (docstring trimmed for brevity - same as before) ...
    """
    # Retrieve the API key from environment variables within the function
    api_key = os.environ.get("OPENWEATHER_API_KEY")

    if not api_key:
        return {"error": "‚ùå Configuration error: OpenWeatherMap API key is not set."}

    if not location or len(location.strip()) < 2:
         return {"error": "‚ùå Input error: Please provide a valid location."}

    try:
        url = (
            f"https://api.openweathermap.org/data/2.5/forecast"
            f"?q={location}&appid={api_key}&units=metric"
        )

        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        if not data or not data.get("list"):
             return {"error": f"‚ùå API returned no data or invalid format for '{location}'. Please check the location spelling."}

        forecast_data = []
        seen_dates = set()

        for entry in data.get("list", []):
            dt_txt = entry.get("dt_txt")
            if not dt_txt:
                continue

            date = dt_txt.split(" ")[0]

            if date not in seen_dates and len(forecast_data) < forecast_days:
                seen_dates.add(date)

                temp_main = entry.get("main", {})
                weather_info = entry.get("weather", [{}])[0]

                forecast_entry = {
                    "date": date,
                    "high": round(temp_main.get("temp_max", 0.0), 1),
                    "low": round(temp_main.get("temp_min", 0.0), 1),
                    "precipChance": round(entry.get("pop", 0.0) * 100),
                    "summary": weather_info.get("description", "N/A").capitalize()
                }
                forecast_data.append(forecast_entry)

            if len(forecast_data) >= forecast_days:
                break

        current_summary = data.get("list", [{}])[0].get("weather", [{}])[0].get("description", "N/A").capitalize()
        current = {"summary": current_summary}

        if not forecast_data:
             return {"error": f"‚ùå No valid forecast entries found for '{location}'. API might not have data."}

        return {
            "current": current,
            "forecast": forecast_data
        }

    except requests.exceptions.HTTPError as e:
        if response.status_code == 401:
            return {"error": f"‚ùå API Key Error: Unauthorized. Please check your OpenWeatherMap API key. Details: {e}"}
        elif response.status_code == 404:
             return {"error": f"‚ùå Location Error: Location '{location}' not found by the API. Please check the spelling. Details: {e}"}
        else:
            return {"error": f"‚ùå HTTP error: Failed to fetch data for '{location}'. Status Code: {response.status_code}. Details: {e}"}
    except requests.exceptions.ConnectionError as e:
         return {"error": f"‚ùå Connection error: Could not connect to the OpenWeatherMap API. Check your internet connection. Details: {e}"}
    except requests.exceptions.Timeout as e:
         return {"error": f"‚ùå Timeout error: The request to the OpenWeatherMap API timed out. Details: {e}"}
    except requests.exceptions.RequestException as e:
        return {"error": f"‚ùå Network or API request error: {e}"}
    except ValueError as e:
         return {"error": f"‚ùå Data processing error: {e}"}
    except KeyError as e:
        return {"error": f"‚ùå Data format error: Missing expected data key: {e}. API response structure might have changed."}
    except Exception as e:
        return {"error": f"‚ùå An unexpected error occurred: {e}"}


# --- Visualisation Functions (Included here for completeness with full code) ---

def create_temperature_visualisation(weather_data, location="Location"):
    """
    Creates and displays a temperature forecast visualisation (High and Low).
    ... (docstring trimmed) ...
    """
    if "error" in weather_data or not weather_data.get("forecast"):
        print("Cannot create temperature visualisation: No valid forecast data.")
        return

    forecast = weather_data["forecast"]
    dates = [day.get("date", "N/A") for day in forecast]
    highs = [day.get("high", 0.0) for day in forecast]
    lows = [day.get("low", 0.0) for day in forecast]

    plot_dates = []
    plot_highs = []
    plot_lows = []
    for i, d in enumerate(dates):
        if d != 'N/A':
            try:
                plot_dates.append(datetime.datetime.strptime(d, "%Y-%m-%d"))
                plot_highs.append(highs[i])
                plot_lows.append(lows[i])
            except ValueError:
                print(f"Warning: Could not parse date '{d}'. Skipping entry for visualisation.")
                continue

    if not plot_dates:
        print("No valid dates found in forecast data for temperature visualisation.")
        return

    plt.figure(figsize=(10, 6))
    plt.plot(plot_dates, plot_highs, marker='o', linestyle='-', color='red', label='High Temp (¬∞C)')
    plt.plot(plot_dates, plot_lows, marker='o', linestyle='-', color='blue', label='Low Temp (¬∞C)')
    plt.title(f"Temperature Forecast ({location})")
    plt.xlabel("Date")
    plt.ylabel("Temperature (¬∞C)")
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(matplotlib.dates.DayLocator())
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


def create_precipitation_visualisation(weather_data, location="Location"):
    """
    Creates and displays a precipitation chance forecast visualisation.
    ... (docstring trimmed) ...
    """
    if "error" in weather_data or not weather_data.get("forecast"):
        print("Cannot create precipitation visualisation: No valid forecast data.")
        return

    forecast = weather_data["forecast"]
    dates = [day.get("date", "N/A") for day in forecast]
    precip_chances = [day.get("precipChance", 0) for day in forecast]

    plot_dates = []
    plot_precip_chances = []
    for i, d in enumerate(dates):
         if d != 'N/A':
            try:
                plot_dates.append(datetime.datetime.strptime(d, "%Y-%m-%d"))
                plot_precip_chances.append(precip_chances[i])
            except ValueError:
                print(f"Warning: Could not parse date '{d}'. Skipping entry for visualisation.")
                continue

    if not plot_dates:
        print("No valid dates found in forecast data for precipitation visualisation.")
        return

    plt.figure(figsize=(10, 6))
    plt.bar(plot_dates, plot_precip_chances, color='skyblue', width=0.6)
    plt.title(f"Precipitation Chance Forecast ({location})")
    plt.xlabel("Date")
    plt.ylabel("Precipitation Chance (%)")
    plt.ylim(0, 100)
    plt.grid(axis='y', linestyle='--')
    plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(matplotlib.dates.DayLocator())
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


def create_combined_visualisation(weather_data, location="Location"):
    """
    Creates and displays a combined plot of temperature, precipitation chance,
    and weather summaries.
    ... (docstring trimmed) ...
    """
    if "error" in weather_data or not weather_data.get("forecast"):
        print("Cannot create combined visualisation: No valid forecast data.")
        return

    forecast = weather_data["forecast"]
    dates = [day.get("date", "N/A") for day in forecast]
    highs = [day.get("high", 0.0) for day in forecast]
    lows = [day.get("low", 0.0) for day in forecast]
    precip = [day.get("precipChance", 0) for day in forecast]
    summaries = [day.get("summary", "N/A") for day in forecast]

    plot_dates = []
    plot_highs = []
    plot_lows = []
    plot_precip = []
    plot_summaries = []

    for i, d in enumerate(dates):
         if d != 'N/A':
            try:
                plot_dates.append(datetime.datetime.strptime(d, "%Y-%m-%d"))
                plot_highs.append(highs[i])
                plot_lows.append(lows[i])
                plot_precip.append(precip[i])
                plot_summaries.append(summaries[i])
            except ValueError:
                print(f"Warning: Could not parse date '{d}'. Skipping entry for combined visualisation.")
                continue

    if not plot_dates:
        print("No valid dates found in forecast data for combined visualisation.")
        return

    plt.figure(figsize=(12, 6))
    plt.plot(plot_dates, plot_highs, marker='o', color='red', label='High Temp (¬∞C)')
    plt.plot(plot_dates, plot_lows, marker='o', color='blue', label='Low Temp (¬∞C)')
    plt.bar(plot_dates, plot_precip, alpha=0.3, color='darkblue', label='Precip Chance (%)', width=0.6)

    for i in range(len(plot_dates)):
        plt.text(plot_dates[i], plot_highs[i] + 2, plot_summaries[i], ha='center', fontsize=9, rotation=45, color='dimgray')

    plt.title(f"üìÖ {len(plot_dates)}-Day Weather Forecast for {location}")
    plt.xlabel("Date")
    plt.ylabel("Temperature (¬∞C) / Precip Chance (%)")
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(matplotlib.dates.DayLocator())
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


# --- Natural Language Processing (NLP) Functions ---
# (Same functions as provided before)

def parse_weather_question(question):
    """
    Parses a user's natural language question to extract location and requested day.
    (Simple rule-based approach)
    ... (docstring trimmed) ...
    """
    question = question.lower()
    location = None
    requested_day = "today"

    location_keywords = ["in ", "for "]
    for keyword in location_keywords:
        if keyword in question:
            parts = question.split(keyword, 1)
            remaining_text = parts[1].strip()
            end_location_markers = ["tomorrow", "today", "weather", "forecast", "in", "for", "on ", "this "]
            location_end_index = len(remaining_text)
            for marker in end_location_markers:
                marker_index = remaining_text.find(marker)
                if marker_index != -1:
                    location_end_index = min(location_end_index, marker_index)

            location = remaining_text[:location_end_index].strip().replace("?", "")
            break

    if location is not None and location.strip() == "":
        location = None

    if "tomorrow" in question:
        requested_day = "tomorrow"
    elif "today" in question:
        requested_day = "today"
    elif "in " in question:
        try:
            parts = question.split("in ", 1)
            if len(parts) > 1:
                 words_after_in = parts[1].split()
                 if len(words_after_in) > 1 and "day" in words_after_in[1]:
                    num_days_str = words_after_in[0]
                    if num_days_str.isdigit():
                        num_days = int(num_days_str)
                        requested_day = f"in {num_days} days"
                    elif num_days_str in ["a", "one"]:
                         requested_day = "tomorrow"
        except Exception:
             pass

    elif "on " in question:
         parts = question.split("on ", 1)
         if len(parts) > 1:
              date_str_candidate = parts[1].strip().replace("?", "")
              try:
                  parsed_date = parse_date(date_str_candidate, fuzzy=True)
                  requested_day = parsed_date.strftime("%Y-%m-%d")
              except ValueError:
                  if date_str_candidate in ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]:
                       requested_day = date_str_candidate
                  else:
                       requested_day = "today" # Default if parsing fails
              except Exception:
                  requested_day = "today"


    return location, requested_day

def generate_weather_response(weather_data, location, requested_day):
    """
    Generates a human-readable weather response based on fetched data and requested day.
    ... (docstring trimmed) ...
    """
    if "error" in weather_data:
        return weather_data["error"]

    forecast = weather_data.get("forecast", [])
    current_summary = weather_data.get("current", {}).get("summary", "N/A")

    # Handle "today"
    if requested_day == "today":
        if forecast:
            today_forecast = forecast[0]
            date = today_forecast.get('date', 'N/A')
            low = today_forecast.get('low', 'N/A')
            high = today_forecast.get('high', 'N/A')
            precip_chance = today_forecast.get('precipChance', 'N/A')
            summary = today_forecast.get('summary', 'N/A')
            response = f"Currently in {location}: {current_summary}. "
            response += f"Today's forecast ({date}): {summary}, with a low of {low}¬∞C and a high of {high}¬∞C. "
            response += f"There is a {precip_chance}% chance of precipitation."
            return response
        else:
            return f"‚ùå Could not retrieve today's forecast data for {location}."

    # Handle "tomorrow"
    elif requested_day == "tomorrow":
        if len(forecast) > 1:
            tomorrow_forecast = forecast[1]
            date = tomorrow_forecast.get('date', 'N/A')
            low = tomorrow_forecast.get('low', 'N/A')
            high = tomorrow_forecast.get('high', 'N/A')
            precip_chance = tomorrow_forecast.get('precipChance', 'N/A')
            summary = tomorrow_forecast.get('summary', 'N/A')
            response = f"Tomorrow's forecast for {location} ({date}): {summary}, with a low of {low}¬∞C and a high of {high}¬∞C. "
            response += f"There is a {precip_chance}% chance of precipitation."
            return response
        else:
            return f"‚ö†Ô∏è I can only provide a {len(forecast)}-day forecast. Tomorrow's weather for {location} is not available."

    # Handle "in X days"
    elif requested_day.startswith("in ") and requested_day.endswith(" days"):
        try:
            parts = requested_day.split(" ")
            if len(parts) == 3:
                num_days = int(parts[1])
                target_day_index = num_days

                if 0 <= target_day_index < len(forecast):
                     target_forecast = forecast[target_day_index]
                     date = target_forecast.get('date', 'N/A')
                     low = target_forecast.get('low', 'N/A')
                     high = target_forecast.get('high', 'N/A')
                     precip_chance = target_forecast.get('precipChance', 'N/A')
                     summary = target_forecast.get('summary', 'N/A')

                     relative_phrase = f"in {num_days} day{'s' if num_days != 1 else ''}"

                     response = f"The forecast for {location} {relative_phrase} ({date}): {summary}, with a low of {low}¬∞C and a high of {high}¬∞C. "
                     response += f"There is a {precip_chance}% chance of precipitation."
                     return response
                elif num_days < 0:
                     return "‚ùì I can only provide forecasts for the future."
                else:
                     max_forecast_days_ahead = len(forecast) - 1
                     return f"‚ö†Ô∏è I can only provide a {len(forecast)}-day forecast. The forecast for {location} {requested_day} is not available. I can forecast up to {max_forecast_days_ahead} day{'s' if max_forecast_days_ahead != 1 else ''} ahead."
            else:
                 return f"‚ùì I couldn't understand '{requested_day}'. Please specify the number of days (e.g., 'in 3 days')."

        except ValueError:
             return f"‚ùì I couldn't understand the number of days in '{requested_day}'. Please specify using digits (e.g., 'in 5 days')."
        except Exception:
            return "‚ùì An unexpected error occurred while processing the requested day."

    # Handle specific date (YYYY-MM-DD)
    elif len(requested_day.split('-')) == 3:
         target_date_str = requested_day
         found_day_forecast = None
         for day in forecast:
             if day.get("date") == target_date_str:
                 found_day_forecast = day
                 break

         if found_day_forecast:
             date = found_day_forecast.get('date', 'N/A')
             low = found_day_forecast.get('low', 'N/A')
             high = found_day_forecast.get('high', 'N/A')
             precip_chance = found_day_forecast.get('precipChance', 'N/A')
             summary = found_day_forecast.get('summary', 'N/A')
             response = f"The forecast for {location} on {date}: {summary}, with a low of {low}¬∞C and a high of {high}¬∞C. "
             response += f"There is a {precip_chance}% chance of precipitation."
             return response
         else:
             if forecast:
                 first_date_str = forecast[0].get("date")
                 last_date_str = forecast[-1].get("date")
                 try:
                     first_date = datetime.datetime.strptime(first_date_str, "%Y-%m-%d").date()
                     last_date = datetime.datetime.strptime(last_date_str, "%Y-%m-%d").date()
                     target_date = datetime.datetime.strptime(target_date_str, "%Y-%m-%d").date()

                     if target_date < first_date:
                          return f"‚ö†Ô∏è I can only provide a forecast starting from {first_date_str}. The date you requested ({target_date_str}) is in the past."
                     elif target_date > last_date:
                          return f"‚ö†Ô∏è I can only provide a forecast up to {last_date_str}. The date you requested ({target_date_str}) is too far in the future."
                     else:
                          return f"‚ùå Could not find the forecast for {location} on {target_date_str} within the available data."
                 except ValueError:
                     return f"‚ùå Could not process date ranges to find forecast for {target_date_str}. Date format issue."
             else:
                 return f"‚ùå No forecast data available to check for the date {target_date_str}."


    # Handle weekday name (e.g., "monday")
    elif requested_day in ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]:
         found_day_forecast = None
         today_obj = datetime.date.today()

         for day in forecast:
             date_str = day.get("date")
             if date_str:
                 try:
                     forecast_date_obj = datetime.datetime.strptime(date_str, "%Y-%m-%d").date()
                     if forecast_date_obj.strftime('%A').lower() == requested_day:
                         found_day_forecast = day
                         break
                 except ValueError:
                      pass

         if found_day_forecast:
             date = found_day_forecast.get('date', 'N/A')
             low = found_day_forecast.get('low', 'N/A')
             high = found_day_forecast.get('high', 'N/A')
             precip_chance = found_day_forecast.get('precipChance', 'N/A')
             summary = found_day_forecast.get('summary', 'N/A')
             response = f"The forecast for {location} on {requested_day.capitalize()} ({date}): {summary}, with a low of {low}¬∞C and a high of {high}¬∞C. "
             response += f"There is a {precip_chance}% chance of precipitation."
             return response
         else:
             today_weekday = today_obj.strftime('%A').lower()
             if requested_day == today_weekday:
                  return generate_weather_response(weather_data, location, "today")

             return f"‚ö†Ô∏è Could not find a future {requested_day.capitalize()} within the {len(forecast)}-day forecast for {location}. Or perhaps you meant 'today'?"


    # Fallback for unhandled day requests
    else:
        return f"‚ùì I couldn't understand the requested day or time '{requested_day}'. Please try rephrasing (e.g., 'tomorrow', 'in 3 days', 'on 2023-10-27', 'Friday')."


    # Default fallback
    return "‚ùì Sorry, I encountered an issue generating the weather response."


# --- Main Application Logic (Interactive with NLP) ---
# This is the part that runs the interaction loop

# Ensure API key was obtained before proceeding with the interaction loop
if api_key:
    print("Welcome to the Weather Bot! Ask me about the weather.")
    print("Examples: 'What is the weather in Paris tomorrow?', 'forecast for London on 2023-11-15', 'weather in New York today'.")
    print("Type 'quit' to exit.")

    while True: # This loop keeps asking for questions until you type 'quit'
        try:
            # Use pyinputplus to get the user's question
            user_question = pyip.inputStr("Your question: ").strip()

            # Check if the user wants to quit
            if user_question.lower() == 'quit':
                print("Goodbye!")
                break # Exit the while loop

            # 1. Parse the user's question to get location and day
            location, requested_day = parse_weather_question(user_question)

            # Check if a location was successfully parsed
            if not location:
                print("‚ùå I couldn't understand the location in your question. Please try again using 'in' or 'for' followed by the city and country (e.g., 'weather in Berlin, Germany').")
                continue # Skip the rest of this loop iteration and ask the next question

            # 2. Get weather data for the extracted location
            # We still need the full 5-day forecast to check specific future days
            weather_data = get_weather_data(location=location, forecast_days=5)

            # 3. Generate a human-readable response based on the fetched data and requested day
            # generate_weather_response handles errors from get_weather_data
            weather_response = generate_weather_response(weather_data, location, requested_day)

            # Print the generated response
            print("\n--- Weather Response ---")
            print(weather_response)
            print("------------------------\n") # Add extra newline for spacing

            # Optional: Offer visualisations if data was successfully fetched and no error occurred
            # Check if weather_data contains forecast data and not an error
            if "error" not in weather_data and weather_data.get('forecast'):
                 try:
                     # Offer visuals for the *full* fetched forecast (up to 5 days)
                     show_visuals = pyip.inputYesNo("Show visualisations for the full 5-day forecast? (yes/no): ")
                     if show_visuals.lower() == 'yes':
                         print("\nCreating visualisations...")
                         # Call visualisation functions with the fetched data and location
                         create_temperature_visualisation(weather_data, location=location)
                         create_precipitation_visualisation(weather_data, location=location)
                         create_combined_visualisation(weather_data, location=location)
                 except Exception as visual_e:
                     print(f"An error occurred while trying to show visualisations: {visual_e}")


        except pyip.TimeoutException:
             print("\nInput timed out. Please try again.")
             continue # Ask for input again
        except Exception as e:
            # Catch any other unexpected errors during the interaction loop
            print(f"\nAn unexpected error occurred during interaction: {e}")
            print("Please try asking your question again.")

else:
    # This message is printed if the API key wasn't obtained at the very beginning
    print("\nWeather bot cannot start because the OpenWeatherMap API key is not available.")
```
